```python
x=[]
y=[]
for i in history_data.iterrows():
    y.append(i[1].tolist()[0:4])
    x.append(i[1].values[4])

c=Kline()
c.add_xaxis(x)
c.add_yaxis("kline", y,itemstyle_opts=opts.ItemStyleOpts(color="pink", color0="black"))
c.set_global_opts(datazoom_opts=[opts.DataZoomOpts(type_="inside"),opts.DataZoomOpts(type_="slider")]
    )
c.render_notebook()
```
